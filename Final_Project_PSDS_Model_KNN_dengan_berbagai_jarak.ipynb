# Model K-Nearest Neighbor (KNN)

Algoritma K-Nearest Neighbor (KNN) adalah algoritma Machine Learning yang melakukan klasifikasi sebuah data didasarkan pada sifat-sifat data didekatnya. Sebuah ungkapan yang bisa membantu dengan mudah memahami ide dasar algoritma ini adalah: "Jika kamu ingin mengetahui sifat seseorang, lihatnya sifat-sifat orang terdekatnya. Sifatnya tidak akan jauh berbeda dari mereka."

Gambar di bawah ini dapat meilustrasikan hal di atas dengan jelas



Data baru yang dilambangkan dengan simbol segitiga biru dapat diklasifikasikan sebagai kelompok kotak hijau atau bintang merah jambu. Jika dilihat berdasarkan kriteria jumlah $k=1$ maka data segitiga biru diklasifikasikan ke dalam kelompok kotak hijau. Sedangkan jika dilihat berdasarkan kriteria jumlah $k=3$ maka data tersebut diklasifikasikan sebagai kelompok bintang merah jambu. Hal yang sama juga dapat diterapkan untuk $k=9$. Sebenarnya kita telah melakukan klasifikasi berdasarkan jarak. Kedekatan yang dimaksud katika $k=1, k=3$, dan $k=9$, kita selalu mengukur kedekatan antara data baru (segitiga biru) dengan data-data berlabel kelompok sebelumnya. Berdasarkan ide dasar di atas, maka diturunkanlah dalam sebuah algoritma untuk menentukan kelompok data dari sebuah inputan data baru (segitiga biru).

## Algoritma KNN

1.  Menentukan parameter k (jumlah tetangga paling dekat)
2.  Menghitung Fungsi Jarak (Manhattan, Minkowski, Chebishev, Bray-Curtiz, Canbera) objek terhadap data training yang diberikan
3.  Mengurutkan hasil no 2 secara ascending (berurutan dari nilai tinggi ke rendah)
4.  Mengumpulkan kategori data (Klasifikasi nearest neighbor berdasarkan nilai k)
5.  Dengan menggunakan kategori nearest neightbor yang paling mayoritas maka dapat diprediksikan kategori objek

# Fungsi Jarak

Perhitungan jarak banyak digunakan dalam menentukan tingkat kemiripan atau tidaknya dua buah vektor. Sehingga metode ini banyak digunakan untuk melakukan pengenalan pola (Wurdianarto et al, 2014). Perbedaan perhitungan jarak terdekat/ jarak kemiripan sangat tepat digunakan untuk menganalisis kelas perbedaan. Perhitungan jarak terdekat/jarak kemiripan menggunakan beberapa nilai matriks biasanya digunakan untuk mengekstrak kemiripan objek data dan dibantu dengan proses klasifikasi menggunakan algoritma yang efiesien. Untuk dataset klasifikasi, beberapa metode perhitungan jarak terdekat/jarak kemiripan adalah Manhattan, Minkowski, Chebishev, Bray-Curtiz, Canbera.

# Evaluasi Model Machine Learning

Mengukur Kinerja Algoritma Klasifikasi dengan Confusion Matrix
* Pengukuran terhadap kinerja suatu sistem klasifikasi merupakan hal yang penting. Kinerja sistem klasifikasi menggambarkan seberapa baik sistem dalam mengklasifikasikan data.
* Confusion matrix merupakan salah satu metode yang dapat digunakan untuk mengukur kinerja suatu metode klasifikasi. 
* Pada dasarnya confusion matrix mengandung informasi yang membandingkan hasil klasifikasi yang dilakukan oleh sistem dengan hasil klasifikasi yang seharusnya 
* Pengukuran kinerja menggunakan confusion matrix, terdapat 4 (empat) istilah adalah :

  1. True Positive (TP)
  2. True Negative (TN)
  3. False Positive (FP) 
  4. False Negative (FN)

TP yaitu jumlah data positif yang terklasifikasi dengan benar oleh sistem.

TN yaitu jumlah data negatif yang terklasifikasi dengan benar oleh sistem.

FN yaitu jumlah data negatif namun terklasifikasi salah oleh sistem.

FP yaitu jumlah data positif namun terklasifikasi salah oleh sistem.



## Tabel Confusion Matrix

|Kelas|Terklasifikasi Positif|Terklasifikasi Negatif|
|-------|-------|-------|
|**Positif**|TP(True Positive)|FN(False Negative)|
|**Negatif**|FP(False Positive)|TN(True Negative)|

$$Akurasi=\frac{TP+TN}{TP+TN+FP+FN}*100%$$

$$Presisi=\frac{TP}{FP+TP}*100%$$

$$Recall=\frac{TP}{FN+TP}*100%$$

* Precision adalah tingkat ketepatan antara informasi yang diminta oleh pengguna dengan jawaban yang diberikan oleh sistem
* Recall adalah tingkat keberhasilan sistem dalam menemukan kembali sebuah informasi
* Accuracy didefinisikan sebagai tingkat kedekatan antara nilai prediksi dengan nilai aktual

### Contoh Penerapan Confusion Matrix

```python
import pandas as pd

df =pd.DataFrame({"y_Aktual": [1,0,0,1,0,1,0,0,1,0,1,0],
                 "y_Prediksi": [1,1,0,1,0,1,1,0,1,0,0,0]})
df
```

```python
confusion_matrix = pd.crosstab(df['y_Aktual'], df['y_Prediksi'], rownames=['Aktual'], colnames=['Prediksi'])
confusion_matrix
```

**Menampilkan confusion matrix menggunakan seaborn**

```python
import seaborn as sn
import matplotlib.pyplot as plt

sn.heatmap(confusion_matrix, annot=True)
plt.show()
```

**Mengukur Akurasi Model**

```python
from sklearn.metrics import accuracy_score
accuracy_score(df['y_Aktual'], df['y_Prediksi'])
```

**Mengukur Recall**

```python
from sklearn.metrics import recall_score
recall_score(df['y_Aktual'], df['y_Prediksi'], average=None)
```

**Mengukur Presisi**

```python
from sklearn.metrics import precision_score
precision_score(df['y_Aktual'], df['y_Prediksi'], average=None)
```

## F1-Score

F1-Score adalah tes untuk mengukur akurasi model

F1-Score adalah rata-rata harmonik dari precision dan recall. Jika didapat nilai maksimum 1 (precision dan recall tersebut merupakan hasil terbaik) dan nilai minimum dari score f1 adalah 0. Secara keseluruhan f1 score merupakan ukuran ketepatan dan kekuatan model.

$$F1-Score=\frac{2*(precision*recall)}{precision+recall}=\frac{2TP}{2TP+FP+FN}$$

**Mengukur F1-Score**

```python
from sklearn.metrics import f1_score
f1_score(df['y_Aktual'], df['y_Prediksi'], average=None)
```

### KNN Python from Scratch dengan Jarak Euclidean

**Membangun model KNN**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X=np.array([[1,1], [2,2], [3,2], [12,11], [11,13], [14,15]])
Y=np.array([0,0,0,1,1,1])
x_baru=np.array([9,10])
```

```python
X
```

```python
Y
```

```python
x_baru
```

**Langkah 1:**

Menentukan parameter k (jumlah tetangga paling dekat)

```python
k=3
```

**Langkah 2:**

Menghitung kuadrat jarak euclidean objek terhadap data training yang diberikan.

```python
def jarak_euclidean(x,y):
    return np.sqrt(sum(pow(a-b, 2) for a, b in zip(x,y)))
```

```python
jarak=[]
for x_i in X:
    jarak_euclid=jarak_euclidean(x_i, x_baru)
    jarak.append(jarak_euclid)
    print(jarak)
print(jarak)
```

**Langkah 3:***

Mengurutkan hasil langkah 2 secara ascending (berurutan dari nilai tinggi ke rendah)

```python
urut=np.sort(jarak)
list_urut=[]
for i in range(len(urut)):
    list_urut.append(urut[i])
print(list_urut)
```

```python
urutan=[]
for i in list_urut:
    urtn=jarak.index(i)
    urutan.append(urtn)
print(urutan)
```

atau dengan sederhana dengan menggunakan np.argsort berikut:

```python
np.argsort(jarak)
```

**Langkah 4:**

Mengumpulkan kategori data (Klasifikai nearest neighbor berdasarkan nilai k)

Kelas data terdekat 3 (k=3)

```python
ambil=urutan[:k]
print(Y[ambil])
```

**Langkah 5:**

Dengan menggunakan kategori nearest neighbor yang paling mayoritas maka dapat diprediksikan kategori objek

```python
from collections import Counter
kelas_paling_banyak = Counter(Y[ambil]).most_common(1)
kelas_paling_banyak
```

```python
kelas_paling_banyak[0][0]
```

```python
from sklearn.model_selection import train_test_split
x_baru = np.array([5.8, 2.9, 5.3, 2.2])
X_latih, X_uji, Y_latih, Y_uji = train_test_split(X, Y, test_size=0.20, random_state=1234)
```

```python
def _knn(x):
    # Menghitung jarak antara x baru dengan semua sampel di data latih
    jarak = [jarak_euclidean(x, x_latih) for x_latih in X_latih]
    
    # Mengurutkan data dan mengambil sejumlah k objek
    k_idx=np.argsort(jarak)[:k]
    
    # Mengambil kelas dari k tetangga terdekat dari data latih
    k_neighbor_labels = [Y_latih[i] for i in k_idx]
    
    # Mengambil kelas yang paling sering muncul
    kelas_paling_banyak = Counter(k_neighbor_labels).most_common(1)
    return kelas_paling_banyak[0][0]

def KNN(X_latih, Y_latih, k, x_baru):
    y_pred=[_knn(x_i) for x_i in x_baru]
    return np.array(y_pred)
```

```python
y_pred=KNN(X_latih, Y_latih, k, X_uji)
y_pred
```

```python
Y_uji
```

**Mendefinisikan Confusion Matrix**

```python
def hitung_tp_tn_fn_fp(y_asli, y_pred):
    tp = sum((y_asli == 1)&(y_pred ==1))
    tn = sum((y_asli == 0)&(y_pred ==0))
    fn = sum((y_asli == 1)&(y_pred ==0))
    fp = sum((y_asli == 0)&(y_pred ==1))
    return tp, tn, fp, fn
```

```python
tp, tn, fp, fn = hitung_tp_tn_fn_fp(Y_uji, y_pred)
print('TP Euclidean:',tp)
print('TN Euclidean:',tn)
print('FN Euclidean:',fp)
print('FP Euclidean:',fn)
```

**Mendefinisikan perhitungan akurasi**

```python
def hitung_akurasi(tp,tn,fn,fp):
    return((tp+tn)*100)/float(tp+tn+fn+fp)
```

```python
akurasi= hitung_akurasi(tp, tn, fn, fp)
akurasi
```

**Mendefinisikan perhitungan presisi**

```python
def hitung_presisi(tp, fp):
    return(tp*100)/float(tp+fp)
```

```python
presisi= hitung_presisi(tp, fp)
presisi
```

**Mendefinisikan perhitungan F1-Score**

```python
def hitung_f1_score(tp, fp, fn):
    return (2*tp)/(2*tp + fp + fn)
```

```python
f1_score = hitung_f1_score(tp, fp, fn)
f1_score
```

Formula Jarak Manhattan
$$D(x,y)=\sum_{i=1}^{k} |x_i - y_i|$$

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X=np.array([[1,1], [2,2], [3,2], [12,11], [11,13], [14,15]])
Y=np.array([0,0,0,1,1,1])
x_baru=np.array([9,10])
```

**Langkah 1:**

Menentukan parameter k (jumlah tetangga paling dekat)

```python
k=3
```

**Langkah 2:**

Menghitung kuadrat jarak minkowski objek terhadap data training yang diberikan

Formula Jarak Minkowski

$$D(x,y)=(\sum_{i=1}^{k}|x_i - y_i|^p)^\frac{1}{p}$$

```python
def akar_ke_n(nilai, akar_n):
    nilai_akar=1/float(akar_n)
    return round(float(nilai)**float(nilai_akar),3)

def minkowski(x,y,p_value):
    return akar_ke_n(sum(pow(abs(a-b), p_value) for a, b in zip(x,y)), p_value)
```

```python
jarak=[]
for x_i in X:
    jarak_minkowski=minkowski(x_i, x_baru, 3)
    jarak.append(jarak_minkowski)
    print(jarak)
print(jarak)
```

## KNN dengan Jarak Chebishev

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X=np.array([[1,1], [2,2], [3,2], [12,11], [11,13], [14,15]])
Y=np.array([0,0,0,1,1,1])
x_baru=np.array([9,10])
```

```python
X
```

```python
Y
```

```python
x_baru
```

**Langkah 1:**

Menentukan parameter k (jumlah tetangga paling dekat)

```python
k=3
```

**Langkah 2:**

Menghitung kuadrat jarak Chebishev objek terhadap data training yang diberikan.

Formula Jarak Chebishev

$$D(x,y)=\max_{i}(|x_i - y_i|)$$

```python
def jarak_chebishev(x,y):
    return max(abs(a-b) for a,b in zip(x,y))
```

```python
jarak=[]
for x_i in X:
    jarak_cheb=jarak_chebishev(x_i, x_baru)
    jarak.append(jarak_cheb)
    print(jarak)
print(jarak)
```

**Langkah 3:**

Formula Jarak Bray-Curtiz

$$
D(x,y)=\frac{\sum_{i=1}^{k}|x_i - y_i|}{\sum_{i=1}^{k}|x_i + y_i|}
$$

```python
def jarak1(x,y): # Perhitungan Jarak Bray_Curtiz
    return sum(abs(a-b) for a,b in zip(x,y))

def jarak2(x,y): # Perhitungan Jarak Bray_Curtiz
    return sum(abs(a+b) for a,b in zip(x,y))
```

```python
def jarak_manhattan(x,y):
    return np.sqrt(sum(abs(a-b) for a, b in zip(x,y)))
```

```python
def _knn_manhattan(x):
    # Menghitung jarak antara x baru dengan semua sampel di data latih
    jarak = [jarak_manhattan(x, x_latih) for x_latih in X_latih]
    
    # Mengurutkan data dan mengambil sejumlah k objek
    k_idx=np.argsort(jarak)[:k]
    
    # Mengambil kelas dari k tetangga terdekat dari data latih
    k_neighbor_labels = [Y_latih[i] for i in k_idx]
    
    # Mengambil kelas yang paling sering muncul
    kelas_paling_banyak = Counter(k_neighbor_labels).most_common(1)
    return kelas_paling_banyak[0][0]

def KNN_manhattan(X_latih, Y_latih, k, x_baru):
    y_pred=[_knn_manhattan(x_i) for x_i in x_baru]
    return np.array(y_pred)
```

**Membuat fungsi KNN dengan jarak Canberra**

```python
import scipy.spatial.distance as dst
```

```python
def _knn_canberra(x):
    # Menghitung jarak antara x baru dengan semua sampel di data latih
    jarak = [dst.canberra(x, x_latih) for x_latih in X_latih]
    # Mengurutkan data dan mengambil sejumlah k objek
    k_idx = np.argsort(jarak)[: k]
    # Mengambil kelas dari k tetangga terdekat dari data latih
    k_neighbor_labels = [Y_latih[i] for i in k_idx]
    # Mengambil kelas yang paling sering muncul
    kelas_paling_banyak = Counter(k_neighbor_labels).most_common(1)
    return kelas_paling_banyak[0][0]

def KNN_canberra(X_latih, y_latih, k, x_baru):
    y_pred= [_knn_canberra(x_i) for x_i in x_baru]
    return np.array(y_pred)
```

## Plot Function

```python
import seaborn as sns
```

```python
def bar_plot(df, evalf):

    # Memilih tema plot
    sns.set_theme(style="whitegrid")

    # Membuat sebuah figure dan sebuah axes
    fig, ax=plt.subplots(figsize=(10,8)) # figsize=(x,y) untuk menentukan ukuran gambar

    # Melakukan bar plot
    ax = sns.barplot(data=df,               # Nama DataFrame
                     y=evalf,               # Data sumbu Y
                     x=df.index             # Data sumbu X

                    )
    for bar in ax.patches:
        ax.annotate(format(bar.get_height(), '.2f'),
                   (bar.get_x() + bar.get_width() / 2,
                    bar.get_height()), ha='center', va='center',
                   size=15, xytext=(0, 8),
                   textcoords='offset points')
    # Mengatur sudut rotasi label stepsize sumbu x
    plt.xticks(rotation=80)

    # Mengganti label sumbu x
    ax.set_xlabel('distance')

    # Mengganti label sumbu y
    ax.set_ylabel('{} score'.format(evalf))

    # Mengatur Judul Gambar
    ax.set_title('Evaluation Score of KNN with Various Distance', fontsize=16, fontweight='bold')

    # Menampilkan Gambar
    plt.show()
```
