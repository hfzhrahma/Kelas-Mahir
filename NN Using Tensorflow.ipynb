## Model Neural Networks Dasar

Tensorflow adalah sebuah library pada python untuk Deep Learning. Pada bagian ini kita akan fokus pada syntaks bagaimana melakukan pemodelan Neural Networks menggunakan library TensorFlow python. Pada bagian ini kita akan mulai pada kasus yang paling sederhana yaitu Tabel Logika "And" sebagai berikut:

|$x_1$|$x_2$|$y$|
|-------|-------|-------|
|$1$|$1$|$1$|
|$1$|$0$|$0$|
|$0$|$1$|$0$|
|$0$|$0$|$0$|

Perhatikan bahwa $x_1$ dan $x_2$ adalah input dan $y$ adalah output yang diinginkan. Maka ukuran input adalah dua kolom dan 4 baris sedangkan input adalag sebuah nilai antara 0 dan 1

```python
# Import Library
import tensorflow as tf
from tensorflow import keras
import numpy as np
```

```python
# Menampilkan versi Tensorflow
tf.__version__
```

Tiga Tahap Penting:
1.  Menyusun Arsitektur Model
2.  Mengkompilasi Model
3.  Training Model

```python
# Menyusun Model
model = keras.Sequential([                         # Membentuk model sequntial
    tf.keras.layers.Dense(units=1,
                         input_shape=[2],
                         activation='relu'),       # Menambah layer pada model NN
])

# Compile Model
model.compile(loss='mean_squared_error',           # Menentukan fungsi loss
             optimizer='sgd')                      # Menentukan fungsi optimizer

model.summary()                                    # Menmpilkan summary model
```

Berdasarkan model summary di atas kita sedang membuat model NN dengan satu titik layer. Kemudian hasil parameter adalah tiga. Itu berarti terdapat tiga parameter, apa saja? Yaitu $w_1$, $w_2$, karena inputna adalah $x_1$ dan $x_2$ serta $b$ yaitu bias. Berikut adalah visualisasi NN modelnya:



```python
# Mendefinisikan Data Input (X) & Output (Y)
x = np.array([[1,1], [0,1], [0,1], [0,0]], dtype="int")
y = np.array([[1], [0], [0], [0]], dtype="int")
```

```python
# Menampilkan data x
x
```

```python
# Menampilkan data y
y
```

```python
# Melatih model (fitting/training)
history = model.fit(x,y, epochs=2000, verbose=2)
```

```python
# Prediksi Model
model.predict(x)
```

```python
# Membulatkan hasil pemodelan
model.predict(x).round()
```

```python
# Menampilkan nilai bobot
model.get_weights()
```
