##Regresi Linear Python Grim Scratch

**Membangun Model Regresi Linear**

Pertama, import libary python yang diperlukan yaitu:
1.   Numpy
2.   Pandas
3.   Matplotlib



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
X_train = np.array([1, 3, 5, 6, 7, 8])
Y_train = np.array([1, 5, 9, 11, 13, 15])
```

```python
#Proses Training Model

#Menghitung rata-rata inputs and outputs
x_rata2 = np.mean(X_train)
print("Rata-rata X yaitu :", x_rata2)

y_rata2 = np.mean(Y_train)
print("Rata-rata Y yaitu :", y_rata2)

#Jumlah data
n = len(X_train)
print("Jumlah data :", n)

#Menuliskan formula untuk menghitung beta 0 dan beta 1
A=0
B=0
for i in range(n):
  a = (X_train[i] - x_rata2) * (Y_train[i] - y_rata2)
  print("(", X_train[i], "-", x_rata2, ")", "*", "(", Y_train[i], "-", y_rata2, ")=", a )
  A += a
  b = (X_train[i]- x_rata2)**2
  print("(", X_train[i], "-", x_rata2, ")", "^2", "=", b)
  B += b

print("Pembilang = ", A)
print("Penyebut = ", B)

#Menampilkan Koefisien Regresi Linear
b_1 = A/B
print("b_1 = ", b_1)

b_0 = y_rata2 - (b_1 * x_rata2)
print("b_0 = ", b_0)
```

```python
#Plot Nilai Data
x_max = np.max(X_train)
x_min = np.min(X_train)

#Garis Regresi
x = np.linspace(x_min, x_max, 1000)
y = b_0 + b_1 * x

#Plot Garis Regresi
plt.plot(x,y, color='red', label='Regresi Linear')

#Plot Titik Data
plt.scatter(X_train, Y_train, color = 'blue', label='Data')

#Label Sumbu X
plt.xlabel('X')

#Label Sumbu Y
plt.ylabel('Y')

#Menampilkan Lagenda
plt.legend()

#Menampilkan Gambar hasil Plot
plt.show()
```

##Implementasi Model Regresi Linear pada Model Data Real

Kita akan menggunakan dataset yang berisi **Nilai Angka 0 sampai 100** dan **Nilai yang dihasilkan menggunakan fungsi Excel NORMINV(RAND(), x, 3)**

```python
data = pd.read_csv('https://raw.githubusercontent.com/hfzhrahma/Kelas-Mahir/main/test.csv')

#Melihat ukuran data
print(data.shape)

#Menampilkan 5 data teratas
data.head()
```

Mencari hubungan antara data X (Lama waktu belajar Matematika) dan data Y (Nilai Ulangan Matematika)

```python
#Mendefinisikan Variabel Input(X) dan Output(Y)
X = data['x'].values.reshape(-1,1)
Y = data['y'].values.reshape(-1,1)
```

```python
#Membagi data menjadi data training dan data testing
X_train, X_test, Y_train, y_test = train_test_split(X,Y, test_size=0.2, random_state=0)
```

```python
#Proses Training Model

#Menghitung rata-rata inputs dan outputs
x_rata2 = np.mean(X_train)
y_rata2 = np.mean(Y_train)

#Jumlah data
n = len(X_train)

#Menuliskan formula untuk menghitung beta 0 dan beta 1
A=0
B=0
for i in range(n):
  A += (X_train[i] - x_rata2) * (Y_train[i] - y_rata2)
  B += (X_train[i]- x_rata2)**2

b_1 = A / B
b_0 = y_rata2 - (b_1 * x_rata2)

#Menampilkan Koefisien Regresi Linear
print(b_0, b_1)
```

Dalam notasi matematika, hasil ini dapat ditulis sebagai berikut:

y = -0.53332421 + 1.0144497 * x

Sekarang kita memiliki model linear. Mari kita plot secara grafis

```python
#Plot Nilai Data
x_max = np.max(X) + 100
x_min = np.min(X) - 100

#Garis Regresi
x = np.linspace(x_min, x_max, 1000)
y = b_0 + b_1 * x

#Plot Garis Regresi
plt.plot(x,y, color='red', label='Regresi Linear')

#Plot Titik Data
plt.scatter(X, Y, color = 'blue', label='Data')

#Label Sumbu X
plt.xlabel('Nilai Matematika')

#Label Sumbu Y
plt.ylabel('Lama Belajar')

#Menampilkan Lagenda
plt.legend()

#Menampilkan Gambar hasil Plot
plt.show()
```

SSE adalah jumlah kuadrat galat sedangkat SST jumlah kuadrat total.

R^2 skor biasanya memiliki nilai antara 0 hingga 1. Nilai ini juga bisa bernilai negatif jika model regresi benar-benar salah. Dibawah ini adalah kode pemprograman untuk menghitung R^2 skor

```python
SST = 0
SSE = 0
y_pred = b_0 + b_1*X_test
y_rata2 = np.mean(y_test)
for i in range(len(X_test)):
  SST += (y_test[i] - y_rata2)**2
  SSE += (y_test[i] - y_pred[i])**2

score = 1-(SSE/SST)
print("Koefisien Determinasi :", score)
```
