## Algoritma K-Nearest Neighbor (KNN)

Algoritma K-Nearest Neighbor (KNN) adalah algoritma Machine Learning yang melakukan klasifikasi sebuah data didasarkan pada sifat-sifat data didekatnya. Sebuah ungkapan yang bisa membantu dengan mudah memahami ide dasar algoritma ini adalah: "Jika kamu ingin mengetahui sifat seseorang, lihatnya sifat-sifat orang terdekatnya. Sifatnya tidak akan jauh berbeda dari mereka."

Gambar di bawah ini dapat meilustrasikan hal di atas dengan jelas



Data baru yang dilambangkan dengan simbol segitiga biru dapat diklasifikasikan sebagai kelompok kotak hijau atau bintang merah jambu. Jika dilihta berdasarkan kriteria jumlah k=1 maka data segitiga biru diklasifikasikan ke dalam kelompok kotak hijau. Sedangkan jika dilihat berdasarkan kriteria jumlah k=3 maka data tersebut diklasifikasikan sebagai kelompok bintang merah jambu. Hal yang sama juga dapat diterapkan untuk k=9. Sebenarnya kita telah melakukan klasifikasi berdasarkan jarak. Kedekatan yang dimaksud katika k=1, k=3, dan k=9, kita selalu mengukur kedekatan antara data baru (segitiga biru) dengan data-data berlabel kelompok sebelumnya. Berdasarkan ide dasar di atas, maka diturunkanlah dalam sebuah algoritma untuk menentukan kelompok data dari sebuah inputan data baru (segitiga biru).

## Algoritma KNN

1.  Menentukan parameter k (jumlah tetangga paling dekat)
2.  Menghitung jarak eucliden objek terhadap data training yang diberikan
3.  Mengurutkan hasil no 2 secara ascending (berurutan dari nilai tinggi ke rendah)
4.  Mengumpulkan kategori data (Klasifikasi nearest neighbor berdasarkan nilai k)
5.  Dengan menggunakan kategori nearest neightbor yang paling mayoritas maka dapat diprediksikan kategori objek

## Contoh Perhitungan K-Nearest Neighbor

Diketahui data $x_1$ dan $x_2$ adalah sebagai berikut:


| $x_1$ | $x_2$ | Kelas |
| --- | --- | --- |
| $$1$$ | $$1$$ | $$0$$ |
| $$2$$ | $$2$$ | $$0$$ |
| $$3$$ | $$2$$ | $$0$$ |
| $$12$$ | $$11$$ | $$1$$ |
| $$11$$ | $$13$$ | $$1$$ |
| $$14$$ | $$15$$ | $$1$$ |



Kita diminta menentukan kelas dari data baru yaitu (9,10) dengan parameter k=3

**Langkah 1:**

Menentukan parameter k (jumlah tetangga paling dekat)

k=3

**Langkah 2:**
Menghitung jarak eucliden objek terhadap data training yang diberikan.

Formula jarak Euclidean:

$$ d_ij = \sqrt[]{(x1_i-x1_j)^2+(x2_i-x2_j)^2} $$ untuk $ i,j = 1,2,3,...,n $

| $x_1$ | $x_2$ | Jarak Euclidean dengan (9,10) |
| --- | --- | --- |
| $$1$$ | $$1$$ | $$\sqrt{(9-1)^2+(10-1)^2}=12.04$$ |
| $$2$$ | $$2$$ | $$\sqrt{(9-2)^2+(10-2)^2}=10.63$$ |
| $$3$$ | $$2$$ | $$\sqrt{(9-3)^2+(10-2)^2}=10$$ |
| $$12$$ | $$11$$ | $$\sqrt{(9-12)^2+(10-11)^2}=3.16$$ |
| $$11$$ | $$13$$ | $$\sqrt{(9-11)^2+(10-13)^2}=3.60$$ |
| $$14$$ | $$15$$ | $$\sqrt{(9-14)^2+(10-15)^2}=7.07$$ |

**Langkah 3:**
Mengurutkan hasil langkah 2 secara ascending (berurutan dari nilai tinggi ke rendah)

| $x_1$ | $x_2$ | Jarak Euclidean dengan (9,10) | Urutan (Jarak Terpendek) |
| --- | --- | --- | --- |
| $$1$$ | $$1$$ | $$\sqrt{(9-1)^2+(10-1)^2}=12.04$$ | $$6$$ |
| $$2$$ | $$2$$ | $$\sqrt{(9-2)^2+(10-2)^2}=10.63$$ | $$5$$ |
| $$3$$ | $$2$$ | $$\sqrt{(9-3)^2+(10-2)^2}=10$$ | $$4$$ |
| $$12$$ | $$11$$ | $$\sqrt{(9-12)^2+(10-11)^2}=3.16$$ | $$1$$ |
| $$11$$ | $$13$$ | $$\sqrt{(9-11)^2+(10-13)^2}=3.60$$ | $$2$$ |
| $$14$$ | $$15$$ | $$\sqrt{(9-14)^2+(10-15)^2}=7.07$$ | $$3$$ |

**Langkah 4:**
Mengumpulkan kategori data (Klasifikasi nearest neighbor berdasarkan nilai k)

Kelas data 3 data terdekat (k=3)

(12,11)=1, (11,13)=1, (14,15)=1

**Langkah 5:**
Dengan menggunakan kategori neasrest neighbor yang paling mayoritas maka dapat diprediksikan ketegorinya objek.

Kelas data untuk k=3

'0'=0, '1'=3

Maka disimpulkan bahwa data (9,10) diklasifikasikan ke kelas '1'

## KNN Python From Scratch

**Membangun Model KNN**

Pertama, import library python yang dibutuhkan yaitu:
1)   Numpy

2)   Pandas

3)   Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

Untuk memahami konsep, kita akan menggunakan data X dan Y seperti yang kita tunjukkan pada proses penghitungan manual

```python
X=np.array([[1,1], [2,2], [3,2], [12,11], [11,13], [14,15]])
Y=np.array([0,0,0,1,1,1])
x_baru=np.array([9,10])
```

```python
X
```

```python
Y
```

```python
x_baru
```

**Langkah 1:**

Menentukan parameter k (jumlah tetangga paling dekat)

```python
k=3
```

**Langkah 2:**

Menghitung kuadrat jarak eucliden objek terhadap data training yang diberikan.

```python
def jarak_euclidean(x,y):
    return np.sqrt(sum(pow(a-b, 2) for a, b in zip(x,y)))
```

```python
jarak=[]
for x_i in X:
    jarak_euclid=jarak_euclidean(x_i, x_baru)
    jarak.append(jarak_euclid)
    print(jarak)
print(jarak)
```

**Langkah 3:**

Mengurutkan hasil langkah 2 secara ascending (berurutan dai nilai tinggi ke rendah)

```python
urut=np.sort(jarak)
list_urut=[]
for i in range(len(urut)):
    list_urut.append(urut[i])
print(list_urut)
```

```python
urutan=[]
for i in list_urut:
    urtn=jarak.index(i)
    urutan.append(urtn)
print(urutan)
```

atau dengan sederhana dapat menggunakan *np.argsort* seperti berikut:

```python
np.argsort(jarak)
```

**Langkah 4:**

Mengumpulkan kategori data (Klasifikasi nearest neightbor berdasarkan nilai k)

Kelas data 3 data data terdekat (k=3)

```python
ambil=urutan[:k]
print(Y[ambil])
```

**Langkah 5:**

Dengan menggunakan kategori nearest neighbor yang paling mayoritas maka dapat diprediksikan kategori objek

```python
from collections import Counter
kelas_paling_banyak = Counter(Y[ambil]).most_common(1)
kelas_paling_banyak
```

```python
kelas_paling_banyak[0][0]
```

## Implementasi Model KNN pada Data Real

Kita akan menggunakan dataset iris

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

data = pd.read_csv('https://raw.githubusercontent.com/hfzhrahma/Kelas-Mahir/main/IRIS.csv')

#Melihat ukuran data
print(data.shape)

#Menampilkan 5 data teratas
data.head()
```

Memilih dataset X(iris, data) dan dataset Y(iris, terget).

```python
data1 = data[data.columns[::-1]]
data2 = data1[data1.columns[1:5]]
data3 = np.asarray(data2)

print(data3)
print(type(data3))
```

Membagi dataset menjadi data latih dan data uji

```python
data.species[data.species == 'Iris-setosa'] = 0
data.species[data.species == 'Iris-versicolor'] = 1
data.species[data.species == 'Iris-virginica'] = 2
data
```

```python
jenis1 = data['species']
jenis = np.array(jenis1)
jenis
```

```python
from sklearn.model_selection import train_test_split
X, Y = data3, jenis
x_baru = np.array([5.8, 2.9, 5.3, 2.2])
X_latih, X_uji, Y_latih, Y_uji = train_test_split(X, Y, test_size=0.20, random_state=1234)
```

Menuliskan algoritma KNN dalam bentuk fungsi

```python
def _knn(x):
    # Menghitung jarak antara x baru dengan semua sampel di data latih
    jarak = [jarak_euclidean(x, x_latih) for x_latih in X_latih]
    
    # Mengurutkan data dan mengambil sejumlah k objek
    k_idx=np.argsort(jarak)[:k]
    
    # Mengambil kelas dari k tetangga terdekat dari data latih
    k_neighbor_labels = [Y_latih[i] for i in k_idx]
    
    # Mengambil kelas yang paling sering muncul
    kelas_paling_banyak = Counter(k_neighbor_labels).most_common(1)
    return kelas_paling_banyak[0][0]

def KNN(X_latih, Y_latih, k, x_baru):
    y_pred=[_knn(x_i) for x_i in x_baru]
    return np.array(y_pred)
```

```python
y_pred=KNN(X_latih, Y_latih, k, X_uji)
y_pred
```

## Evaluasi Model

Evaluasi model digunakan untuk melihat seberapa banyak baik metode ini bekerja pada data uji yang diberikan. Dalah satu metrik yang dapat digunakan adalah akurasi. Metriks ini menghitung jumlah hasil prediksi yang sesuai dengan label sesungguhnya kemudian dibagi dengan jumlah keseluruhan label. Menggunakan bahasa python akurasi dapat dituliskan sebagai berikut.

```python
def akurasi(y_asli, y_pred):
    akurasi = np.sum(y_asli == y_pred) / len(y_asli)
    return akurasi
```

```python
akurasi(Y_uji,y_pred)
```

Diperoleh hasil akurasi $1.0$ artinya metode KNN dapat melakukan prediksi dengan tepat untuk semua data uji yang digunakan.
